# merging the set and themes data sets

```python
sets_themes=pd.merge(sets,themes,how="left",left_on = "theme_id",right_on ="id",suffixes=("_sets","-themes")) 
sets_themes
```

# Number of parts by themes

```python
Num_of_parts_themes=sets_themes.groupby("name-themes").sum().reset_index()
Num_of_parts_themes=Num_of_parts_themes[["name-themes","num_parts"]].sort_values('num_parts', ascending= False)
Num_of_parts_themes
```

```python
num_parts_year = sets_themes.groupby("year").sum().reset_index()
num_parts_year= num_parts_year[["year","num_parts"]].sort_values("num_parts", ascending = False)
num_parts_year
```

```python
unique_years = sets_themes["year"].unique()
unique_years.sort()
unique_years
```

```python
sets_themes["year"] > 2000
sets_themes["century"] = np.where(sets_themes["year"]>2000, "21st_century", "20th_century")
sets_themes
```

# Number of sets in the 20th and 21st century

```python
sets_per_century = sets_themes.groupby("century").count().reset_index()
sets_per_century[["century","set_num"]]
```

```python
century_21st = sets_themes[sets_themes["century"] =="21st_century"]
train= century_21st[century_21st['name-themes'].str.contains('Train')]
train
```

```python
popular_theme_per_year = century_21st.groupby(['year','name-themes']).count().reset_index()
popular_theme_per_year = popular_theme_per_year[['year','name-themes',"set_num"]].sort_values('set_num', ascending = False)
popular_theme_per_year.drop_duplicates('year').sort_values('set_num', ascending = False)
```

```python
colors= pd.read_csv('colors.csv')
colors
```

```python
inventory_parts= pd.read_csv('Inventory_parts.csv')
inventory_parts
```

```python
Color_inventory = pd.merge(inventory_parts,colors,how="left",left_on = "color_id",right_on = "id")
Color_inventory 
```

```python
Color_inventory.rename(columns= {'name':'color_name'}, inplace= True)
Color_inventory
```

# Most popular colors

```python
Popular_color = Color_inventory.groupby('color_name').sum().reset_index()
Popular_color[['color_name',"quantity"]].sort_values('quantity', ascending = False)
```
